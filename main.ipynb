```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
cols = [
    'Entity', 'Code', 'Year', 'Alochol use',
    'Diet low in fruits', 'Unsafe water source','Secondhand smoke',
    'Smoking','Air pollution','High body mass index','Drug use'
]

df = pd.read_csv('dataset.csv', usecols=cols)

paises = ['Brazil', 'Pakistan', 'United States', 'Japan', 'Russia', 'Mexico', 'Ethiopia', 'Philippines','India','China']

dataframe = df[df['Entity'].isin(paises)]
```

```python
causas = list(filter(lambda col: col not in ['Entity', 'Code', 'Year'], cols))
```

```python
paises = dataframe['Entity'].unique()
cores_paises = {pais: np.random.rand(3,) for pais in paises}

for causa in causas:
    plt.figure(figsize=(12, 6)) 
    
    for pais in paises:
        dataframe_pais = dataframe[dataframe['Entity'] == pais]
        plt.bar(pais, dataframe_pais[causa], label=pais, color=cores_paises[pais])
    
    plt.xlabel('País')
    plt.ylabel('Número de Mortes')
    plt.xticks(rotation=90)  
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
    
    plt.grid(True)  
    plt.title(f'Número de Mortes por Causa: {causa}')
    plt.tight_layout() 
    plt.show()
```

## Conclusões: Número de Mortes (Amostra Global)
- Dentre os países da amostra (Brasil, China, Etiopia, India, Japão, México, Paquistão, Filipinas, Russia, EUA) podemos obter as seguintes informações:
    - India e China (países atualmente com maior população mundial) possuem os maiores números de morte em todos os aspectos.
    - Mesmo com uma população menor, os Estados Unidos apresenta um alto número de casos de morte por drogas, principalmente em comparação aos dois gigantes populacionais (China e India), o que mostra um grave problema de saúde nacional americana:
        - China: Aproximadamente 1,5 milhões de casos;
        - India: Aproximadamente 78 mil casos;
        - Estados Unidos: Quase 1,3 milhões de casos.
    - Da mesma forma, a Russia apresenta uma elevada taxa de mortes por alcolismo, sendo o terceiro país com maior índice, contando com mais de 300 mil casos, se aproximando da China (segunda colocada).

```python
df_brasil = dataframe[dataframe['Entity'] == 'Brazil']

plt.figure(figsize=(12, 6))

for causa in causas:
    plt.plot(df_brasil['Year'], df_brasil[causa], label=causa)

plt.xlabel('Ano')
plt.ylabel('Número de Mortes')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

plt.grid(False)  
plt.title('Número de Mortes por Causa no Brasil')
plt.show()
```

## Conclusões: Número de Mortes por Causa no Brasil
- O número de mortes no Brasil por tabagismo sempre foi o maior em comparação ao restante dos casos em questão (apresentando uma média de 160 mil mortes ao longo de 30 anos), entretanto, nos últimos anos, casos de morte por obesidade dispararam e ultrapassaram essa marca, registrando mais de 175 mil casos no ano de 2019;
    - Segundo (CAMPOS, 2019), o excesso de peso e obesidade causam 168 mil mortes por ano no Brasil;
- Em contra partida, o número de mortes por má qualidade da água, caiu drasticamente de 1990 para os anos atuais, o que pode indicar uma melhoria no quadro de saneamento básico nacional, a medida que as pessoas passam a ter acesso á água potável e de boa qualidade.
